```python
def get_redundant_pairs(s):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = s.columns
    for i in range(0, s.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(s, n=5):
    labels_to_drop = get_redundant_pairs(s)
    au_corr = s.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]
```
